```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko)\
    Chrome/91.0.4472.77 Safari/537.36',
}

count=0
for i in range(0,len(allLinks)):

    foil_list=[]
    
    link=allLinks[i]

    response = requests.get(link,headers=headers,verify=False)
    doc=BeautifulSoup(response.text,'html.parser')

    foil_dict={}
    foil_dict['url']=link

    count=count+1
    print(count,"=", link)


    container=doc.find("div",class_="wrapper")

    main_container=container.find(class_="request-info")
    foil_dict['foil_id']=main_container.find('h3').text.strip()
    foil_dict['title']=main_container.find(class_="request-title-text lead").text.strip()
    foil_dict["agency_request_summary"]=main_container.find(class_="agency-request-summary-text lead").text.strip()

    left_container=container.find("div",class_="container-fluid")
    foil_dict['status']=left_container.find_all(['h4','span'])[0].text.strip()
    foil_dict['ass_agency']=left_container.find_all(['h4','span'])[1].text.strip()
    foil_dict['due_date']=pd.to_datetime(left_container.find_all(['h4','span'])[2].text.strip())
    foil_list.append(foil_dict)

    df=pd.DataFrame(foil_list)
    df.to_csv(f'data-files/{allIds[i]}.csv', index=False)
    
```
